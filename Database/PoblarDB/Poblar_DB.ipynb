# **Direcciones**

```python
import random

departamentos_y_ciudades = {
    "Antioquia": ["Medellín", "Envigado", "Itagüí"],
    "Cundinamarca": ["Bogotá", "Soacha", "Chía"],
    "Valle del Cauca": ["Cali", "Palmira", "Tuluá"],
    "Atlántico": ["Barranquilla", "Soledad", "Malambo"],
    "Santander": ["Bucaramanga", "Floridablanca", "Giron"],
    "Bolívar": ["Cartagena", "Turbaco", "Arjona"]
}

direcciones = {}

for id_usuario in range(1, 101):
    region = random.choice(list(departamentos_y_ciudades.keys()))
    ciudad = random.choice(departamentos_y_ciudades[region])
    codigo_postal = random.randint(110001, 599999)
    calle = random.randint(1, 150)
    carrera = random.randint(1, 100)
    numero = f"{calle}#{carrera}-{random.randint(1, 99)}"

    direcciones[id_usuario] = {
        "id_usuario": id_usuario,
        "pais": "Colombia",
        "region": region,
        "ciudad": ciudad,
        "codigo_postal": codigo_postal,
        "direccion": f"Calle {calle} # {carrera}-{random.randint(1, 99)}"
    }

#Exportar ventas en el futuro si es necesario
with open('direcciones.json', 'w', encoding='utf-8') as f:
    json.dump(direcciones, f, ensure_ascii=False, indent=4)
```

## **Función para obtener una fecha aleatoria en un rango de tiempo**

```python
import random
from datetime import datetime, timedelta

def fecha_aleatoria(anio_inicio: int, anio_fin: int) -> str:
    """
    Genera una fecha aleatoria en formato "YYYY-MM-DD HH:MM:SS.sss"
    dentro del rango de años [anio_inicio, anio_fin], sin superar el presente.

    Args:
        anio_inicio (int): Año inicial (inclusive).
        anio_fin (int): Año final (inclusive).

    Returns:
        str: Fecha aleatoria formateada.
    """
    ahora = datetime.now()
    # Ajustar el rango para que no supere el año actual
    anio_fin = min(anio_fin, ahora.year)

    # Fecha mínima y máxima en el rango dado
    fecha_min = datetime(anio_inicio, 1, 1)
    fecha_max = datetime(anio_fin, 12, 31, 23, 59, 59, 999999)

    # No permitir que la fecha máxima supere la fecha actual
    if fecha_max > ahora:
        fecha_max = ahora

    # Diferencia total en segundos y microsegundos
    delta = fecha_max - fecha_min
    segundos_totales = delta.total_seconds()

    # Generar un offset aleatorio en segundos (float para incluir microsegundos)
    segundos_aleatorios = random.uniform(0, segundos_totales)

    fecha_random = fecha_min + timedelta(seconds=segundos_aleatorios)

    # Formatear con milisegundos
    fecha_formateada = fecha_random.strftime("%Y-%m-%d %H:%M:%S.%f")[:-3]

    return fecha_formateada
```
